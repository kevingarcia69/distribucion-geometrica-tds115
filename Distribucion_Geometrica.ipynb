```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p = 0.3
k = np.arange(1, 15)

probabilidades = geom.pmf(k, p)

for i, prob in zip(k, probabilidades):
    print(f"P(X = {i}) = {prob:.4f}")

plt.figure()
plt.stem(k, probabilidades)
plt.title("Distribución Geométrica (p = 0.3)")
plt.xlabel("Número de intentos")
plt.ylabel("Probabilidad")
plt.grid()

plt.show()
```
